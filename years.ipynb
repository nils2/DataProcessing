```python
# Load library
import numpy as np
import pandas as pd

from bokeh.plotting import figure
from bokeh.io import output_file, show


import csv
```

```python
#load dataframe
df = pd.read_csv("datagun.csv")
df
```

```python
#count number of incidents for each year
year_counter = 0
years = [2014,2015,2016,2017,2018] 
incident_counts = []
scaled_counts = []

#count incidents for every year
for j in years:            
    for i in df["year"]:
        if i == j:
            year_counter += 1
    
    #append incident counts to list
    incident_counts.append(year_counter)
    year_counter = 0
    
#divide by 10000 for scaling
for i in incident_counts:
    i = i / 10000
    scaled_counts.append(i)
    
print(scaled_counts)
```

```python
#draw bar chart for incident counts

output_file("years.html")

str_years = ['2014','2015','2016','2017','2018']

p = figure(x_range=str_years, plot_height=250, title="Incident Counts divided by 10000 for each Year",
           toolbar_location=None, tools="")

p.vbar(x=str_years, top=scaled_counts, width=0.5)

p.xgrid.grid_line_color = None
p.y_range.start = 0


show(p)
```

```python
#monthly incident counts

incident_counter_2014 = []
incident_counter_2015 = []
incident_counter_2016 = []
incident_counter_2017 = []
incident_counter_2018 = []

#divide list of incidents into years
with open('datagun.csv') as data:
    reader = csv.reader(data)
    for c, row in enumerate(reader):
        if row[3] == "2014":
            incident_counter_2014.append(row[2])
        if row[3] == "2015":
            incident_counter_2015.append(row[2])
        if row[3] == "2016":
            incident_counter_2016.append(row[2])
        if row[3] == "2017":
            incident_counter_2017.append(row[2])
        if row[3] == "2018":
            incident_counter_2018.append(row[2])

month_counter = 1 
monthly_counts = []

#iterate over list of incidents in a year
for i in range(1,(len(incident_counter_2014))):                             
    
    #check if still the same month
    #if so, increment incident counter
    if incident_counter_2014[i] == incident_counter_2014[i-1]:
        month_counter += 1 
    else:    
        monthly_counts.append(month_counter)                    
        month_counter = 1
    
    #if last incident has been reached, append and reset incident counter
    if i == len(incident_counter_2014)-1:
        monthly_counts.append(month_counter)
        month_counter = 1

#apply to all years
for i in range(1,(len(incident_counter_2015))):                             
    if incident_counter_2015[i] == incident_counter_2015[i-1]:
        month_counter += 1 
    else:    
        monthly_counts.append(month_counter)                    
        month_counter = 1
    if i == len(incident_counter_2015)-1:
        monthly_counts.append(month_counter)
        month_counter = 1
        
for i in range(1,(len(incident_counter_2016))):                             
    if incident_counter_2016[i] == incident_counter_2016[i-1]:
        month_counter += 1 
    else:    
        monthly_counts.append(month_counter)                    
        month_counter = 1
    if i == len(incident_counter_2016)-1:
        monthly_counts.append(month_counter)
        month_counter = 1
        
for i in range(1,(len(incident_counter_2017))):                             
    if incident_counter_2017[i] == incident_counter_2017[i-1]:
        month_counter += 1 
    else:    
        monthly_counts.append(month_counter)                    
        month_counter = 1
    if i == len(incident_counter_2017)-1:
        monthly_counts.append(month_counter)
        month_counter = 1
        
for i in range(1,(len(incident_counter_2018))):                             
    if incident_counter_2018[i] == incident_counter_2018[i-1]:
        month_counter += 1 
    else:    
        monthly_counts.append(month_counter)                    
        month_counter = 1
    if i == len(incident_counter_2018)-1:
        monthly_counts.append(month_counter)
        month_counter = 1
    
        
print(monthly_counts)
print(len(monthly_counts))
      
            
            
        
        
        
    
```
