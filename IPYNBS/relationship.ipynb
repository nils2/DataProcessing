```python
# Load library
import numpy as np
import pandas as pd
from collections import Counter

from bokeh.io import show, output_file
from bokeh.plotting import figure
from bokeh.models import HoverTool

import csv

import matplotlib.ticker as ticker
import matplotlib.cm as cm
import matplotlib as mpl
from matplotlib.gridspec import GridSpec

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datagun.csv", sep=',')

gang = df['participant_relationship'].str.contains("Gang vs Gang").sum()
friends = df['participant_relationship'].str.contains("Friends").sum()
unknown = df['participant_relationship'].str.contains("Unknown").sum()
lovers = df['participant_relationship'].str.contains("Significant others").sum()
family = df['participant_relationship'].str.contains("Family").sum()
invasion = df['participant_relationship'].str.contains("Home Invasion").sum()
robbery = df['participant_relationship'].str.contains("Armed Robbery").sum()
aquaintance = df['participant_relationship'].str.contains("Aquaintance").sum()
neighbor = df['participant_relationship'].str.contains("Neighbor").sum()
co = df['participant_relationship'].str.contains("Co-worker").sum()
drive = df['participant_relationship'].str.contains("Drive by").sum()
mass = df['participant_relationship'].str.contains("Mass").sum()

guns = ['Unknown', 'Gang', 'Friends', 'Family', 'Lovers/Ex', 'Home invasion', 'Armed Robbery', 'Aquaintance', 'Neighbor', 'Co-worker', 'Drive by', 'Mass shooting']
fracs = [unknown, gang, friends, family, lovers, invasion, robbery, aquaintance, neighbor, co , drive, mass]
colors = ['lightgrey', 'red', 'blue', 'yellow', 'pink', 'cyan', 'purple', 'lime', 'green', 'maroon', 'teal', 'black']

# Make square figures and axes
plt.figure(1, figsize=(8,8))

patches, texts = plt.pie(fracs, colors=colors, startangle=90)

plt.suptitle('Relationships between people involved', fontsize=20)
plt.legend(patches, guns, loc="lower left")

plt.axis('equal')

plt.show()
```

```python
df = pd.read_csv("datagun.csv", sep=',')

gang = df['participant_relationship'].str.contains("Gang vs Gang").sum()
friends = df['participant_relationship'].str.contains("Friends").sum()
unknown = df['participant_relationship'].str.contains("Unknown").sum()
lovers = df['participant_relationship'].str.contains("Significant others").sum()
family = df['participant_relationship'].str.contains("Family").sum()
invasion = df['participant_relationship'].str.contains("Home Invasion").sum()
robbery = df['participant_relationship'].str.contains("Armed Robbery").sum()
aquaintance = df['participant_relationship'].str.contains("Aquaintance").sum()
neighbor = df['participant_relationship'].str.contains("Neighbor").sum()
co = df['participant_relationship'].str.contains("Co-worker").sum()
drive = df['participant_relationship'].str.contains("Drive by").sum()
mass = df['participant_relationship'].str.contains("Mass").sum()

output_file("HTML/bars-relationship.html")

guns = ['Gang', 'Friends', 'Family', 'Lovers/Ex', 'Home invasion', 'Armed Robbery', 'Aquaintance', 'Neighbor', 'Co-worker', 'Drive by', 'Mass shooting']
colors = ['red', 'blue', 'yellow', 'pink', 'cyan', 'purple', 'lime', 'green', 'maroon', 'teal', 'black']
hover_tool= HoverTool(tooltips=[('value', '@x'), ('amount', '@top')], mode='vline')

p = figure(x_range=guns, plot_height=500, plot_width=1000, toolbar_location='right', title="Known relationships of the people involved",
           tools=[hover_tool,"pan,wheel_zoom,reset,save"])

p.vbar(x=guns, top=[gang, friends, family, lovers, invasion, robbery, aquaintance, neighbor, co, drive, mass], width=0.9, color=colors)

p.left[0].formatter.use_scientific = False
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.xaxis.axis_label = "Known relation between people involved"
p.yaxis.axis_label = "Amount of persons"

show(p)
```
