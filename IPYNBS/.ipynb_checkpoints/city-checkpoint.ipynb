### Removes all the duplicate cities from the list. So you can see in how many different cities there is gunviolence.

```python
import csv
import pandas as pd

# reads data from the gun violence dataset
reader = pd.read_csv('../DATA/datagun.csv')
city_l = []

# appends city to list and removes duplicates
for city in reader['city_or_county']:
    city = str(city)
    city_l.append(city)
remove = list(set(city_l))

# sorts the cities in alphabetical order 
remove.sort()
print(remove)
print('---------------------------------------------------------------------------------------------------------')
# prints how many different cities are in the list
print (len(remove))
```

### Puts all the cities in a list and searches for the most common cities in the list. This way you can look at which  cities have the most gunviolence. 

```python
import csv
import pandas as pd
from collections import Counter

reader = pd.read_csv('../DATA/datagun.csv')
common_l = []
# appends all the cities in dataset to list
for city in reader['city_or_county']:
    city = str(city)
    common_l.append(city)

# counts the occurence of every city in list
cnt = Counter(common_l)
# the value represent the number of most common cities in list
print(cnt.most_common(100))
```

### Puts the cities and the number of gun incidents into separate lists

```python
import csv
import pandas as pd
from collections import Counter

common_l = []

# appends all the cities in dataset to list
for city in reader['city_or_county']:
    city = str(city)
    common_l.append(city)

# renames all Chicago districts to 'Chicago'
common_l = [city.replace('Chicago (Burnham)', 'Chicago').replace('Chicago (Englewood)', 'Chicago').replace('Chicago (Calumet Park)','Chicago').replace('Chicago (Chicago Lawn)','Chicago').replace('Chicago (Hegewisch)', 'Chicago').replace('Chicago (Lakeview)', 'Chicago').replace('Chicago (Lincoln Park)', 'Chiacgo').replace( 'Chicago (Norridge)', 'Chicago').replace('Chicago (Roseland)', 'Chicago').replace('Chicago Heights', 'Chicago').replace( 'Chicago Heights (Ford Heights)','Chicago') for city in common_l]

common_l.sort()

#print(common_l)
cnt = Counter(common_l)

# change the value for how many cities you want in your result
most_common = cnt.most_common(100)

l = []
# appends cities and their occurence number to list as separate strings
for x in most_common:
    for y in x:
        l.append(y)

# cities got all odd strings (city names) from l in a list
cities =l[::2]
# numbers got all even strings (occurence of cities) from l in a list
numbers = l[1::2]


print(cities)
print('------------------------------------------------------------------')
print(numbers)
```

### Makes bar plot for every gun incident in the cities for the cities with the most incidents

```python
# load library
import numpy as np
import pandas as pd
from collections import Counter

from bokeh.io import show, output_file
from bokeh.plotting import figure
from bokeh.embed import components
from bokeh.models import Axis
from bokeh.palettes import Spectral3, Spectral4
from bokeh.models import HoverTool

import csv
```

```python
# read dataset
df = pd.read_csv("../DATA/datagun.csv", sep=',')
# output name 
output_file("../HTML/bars-cities.html")

hover_tool= HoverTool(tooltips=[('value', '@x'), ('amount', '@top')], mode='vline')

# cities on x-axis and number of gun incidents in city on y-axis
p = figure(x_range=cities, plot_height=600, plot_width=1500, title="Number of gun incidents in cities",
           toolbar_location='right', tools=[hover_tool,"pan,wheel_zoom,reset,save"])

p.vbar(x=cities, top=numbers, width=0.9)
# name of city not on x-axis only hover tool
p.xaxis.visible=None
p.left[0].formatter.use_scientific = False
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.xaxis.axis_label = 'Cities'
p.yaxis.axis_label = 'Number of gun incidents'
show(p)
```
