```python
# Load library
import numpy as np
import pandas as pd

from bokeh.plotting import figure
from bokeh.io import output_file, show
from bokeh.models import HoverTool


import csv
```

```python
#load dataframe
df = pd.read_csv("../DATA/datagun.csv")
df
```

```python
#count number of incidents for each year
year_counter = 0
years = [2014,2015,2016,2017,2018] 
incident_counts = []
scaled_counts = []

#count incidents for every year
for j in years:            
    for i in df["year"]:
        if i == j:
            year_counter += 1
    
    #append incident counts to list
    incident_counts.append(year_counter)
    year_counter = 0
    
#divide by 10000 for scaling
for i in incident_counts:
    i = i / 10000
    scaled_counts.append(i)
    
print(scaled_counts)
```

```python
#draw bar chart for incident counts

output_file("../HTML/years.html")

str_years = ['2014','2015','2016','2017','2018']
hover_tool= HoverTool(tooltips=[('value', '@x'), ('amount', '@top')], mode='vline')

p = figure(x_range=str_years, plot_height=300, title="Incident Counts",
           toolbar_location='right', tools=[hover_tool,"pan,wheel_zoom,reset,save"])

p.vbar(x=str_years, top=scaled_counts, width=0.5)

p.x_range.range_padding = 0.0001
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.xaxis.axis_label = "Years"
p.yaxis.axis_label = "Amount of occurences x10000"


show(p)
```

```python
#monthly incident counts
#dit kan nog korter ws maarja

incident_counter_2014 = []
incident_counter_2015 = []
incident_counter_2016 = []
incident_counter_2017 = []
incident_counter_2018 = []

#divide list of incidents into years
with open('../DATA/datagun.csv') as data:
    reader = csv.reader(data)
    for c, row in enumerate(reader):
        if row[3] == "2014":
            incident_counter_2014.append(row[2])
        if row[3] == "2015":
            incident_counter_2015.append(row[2])
        if row[3] == "2016":
            incident_counter_2016.append(row[2])
        if row[3] == "2017":
            incident_counter_2017.append(row[2])
        if row[3] == "2018":
            incident_counter_2018.append(row[2])

month_counter = 1 
monthly_counts = []

#iterate over list of incidents in a year
for i in range(1,(len(incident_counter_2014))):                             
    
    #check if still the same month
    #if so, increment incident counter
    if incident_counter_2014[i] == incident_counter_2014[i-1]:
        month_counter += 1 
    else:    
        monthly_counts.append(month_counter)                    
        month_counter = 1
    
    #if last incident has been reached, append and reset incident counter
    if i == len(incident_counter_2014)-1:
        monthly_counts.append(month_counter)
        month_counter = 1

#apply to all years
for i in range(1,(len(incident_counter_2015))):                             
    if incident_counter_2015[i] == incident_counter_2015[i-1]:
        month_counter += 1 
    else:    
        monthly_counts.append(month_counter)                    
        month_counter = 1
    if i == len(incident_counter_2015)-1:
        monthly_counts.append(month_counter)
        month_counter = 1
        
for i in range(1,(len(incident_counter_2016))):                             
    if incident_counter_2016[i] == incident_counter_2016[i-1]:
        month_counter += 1 
    else:    
        monthly_counts.append(month_counter)                    
        month_counter = 1
    if i == len(incident_counter_2016)-1:
        monthly_counts.append(month_counter)
        month_counter = 1
        
for i in range(1,(len(incident_counter_2017))):                             
    if incident_counter_2017[i] == incident_counter_2017[i-1]:
        month_counter += 1 
    else:    
        monthly_counts.append(month_counter)                    
        month_counter = 1
    if i == len(incident_counter_2017)-1:
        monthly_counts.append(month_counter)
        month_counter = 1
        
for i in range(1,(len(incident_counter_2018))):                             
    if incident_counter_2018[i] == incident_counter_2018[i-1]:
        month_counter += 1 
    else:    
        monthly_counts.append(month_counter)                    
        month_counter = 1
    if i == len(incident_counter_2018)-1:
        monthly_counts.append(month_counter)
        month_counter = 1
        
        
#divide by 1000 for scaling
scaled_mcounts = []

for i in monthly_counts:
    i = i / 1000
    scaled_mcounts.append(i)
    
print(scaled_mcounts)
print(len(scaled_mcounts))
```

```python
#draw bar chart for monthly incident counts

output_file("../HTML/months.html")

str_months = ['jan-14', 'feb-14', 'mar-14', 'apr-14', 'may-14', 'jun-14','jul-14','aug-14','sep-14','okt-14','nov-14','dec-14',
              'jan-15', 'feb-15', 'mar-15', 'apr-15', 'may-15', 'jun-15','jul-15','aug-15','sep-15','okt-15','nov-15','dec-15',
             'jan-16', 'feb-16', 'mar-16', 'apr-16', 'may-16', 'jun-16','jul-16','aug-16','sep-16','okt-16','nov-16','dec-16',
             'jan-17', 'feb-17', 'mar-17', 'apr-17', 'may-17', 'jun-17','jul-17','aug-17','sep-17','okt-17','nov-17','dec-17',
             'jan-18','feb-18','mar-18']
hover_tool= HoverTool(tooltips=[('value', '@x'), ('amount', '@top')], mode='vline')

p = figure(x_range=str_months, plot_height=300, title="Monthly Incident Counts",
           toolbar_location='right', tools=[hover_tool,"pan,wheel_zoom,reset,save"])

p.vbar(x=str_months, top=scaled_mcounts, width=0.5)

p.x_range.range_padding = 0.0001
p.xaxis.major_label_orientation = 1
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.xaxis.axis_label = "Month with correspoding year"
p.yaxis.axis_label = "Amount of occurences x1000"

show(p)
```
